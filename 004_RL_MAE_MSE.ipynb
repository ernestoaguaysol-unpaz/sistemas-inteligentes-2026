```python
# Librerías necesarias
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Carga del dataset y creación de X e y
dataset = pd.read_csv(‘dataset_hs_jugadas_puntaje.csv')
X = dataset.iloc[:, :-1]# Todas las filas, todas las columnas menos la última
y = dataset.iloc[:, -1] # Todas las filas, última columna
```

```python

# Crear H - objeto de la clase base: Regresión lineal. Modelo sin ajustar / sin entrenar.
modelo = LinearRegression()

# Ajustar H → crear h → ajustar/entrenar el modelo
modelo.fit(X, y)
```

```python
# Obtenemos las predicciones para poder calcular MAE y MSE con el conjunto de entrenamiento
y_pred = modelo.predict(X)

# Calculamos MAE y MSE
mean_absolute_error(y, y_pred), mean_squared_error(y, y_pred)
(547.0700634276697, 419356.9380968478)
```
