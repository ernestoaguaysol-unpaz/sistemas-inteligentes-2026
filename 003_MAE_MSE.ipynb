```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
y_ejemplo = [115, 150, 210, 280]
y_pred_1 = [115, 149, 203, 280]
y_pred_2 = [110, 150, 279, 280]
```

```python
mean_absolute_error(y_ejemplo, y_pred_1), mean_absolute_error(y_ejemplo, y_pred_2)
```

```python
mean_squared_error(y_ejemplo, y_pred_1), mean_squared_error(y_ejemplo, y_pred_2)
```
