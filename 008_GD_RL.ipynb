```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([[0], [10]])
y = np.array([[0], [5]])
N = 2 # X.shape[0]
w0, w1, alpha = -4, -6, 0.0001
historial_MSE = []
```

```python
for i in range(5000):

    suma_para_w0 = 0 # Suma (y - _pred) para todos los ejemplos
    suma_para_w1 = 0 # Suma (x * (y - _pred)) para todos los ejemplos
    MSE = 0

    for j in range(0, N):
        y_pred = w0 + w1 * X[j][0]

        suma_para_w0 += (y[j] - y_pred)
        suma_para_w1 += (y[j] - y_pred) * X[j][0]

        L2 = (y[j] - y_pred) ** 2
        MSE += L2

    w0 = w0 - alpha * (-2/N) * suma_para_w0
    w1 = w1 - alpha * (-2/N) * suma_para_w1

    historial_MSE.append(MSE)
```

```python
plt.scatter(x=X[:, 0],y=y)
plt.axline((0, w0[0]), slope=w1[0], color='C0')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(historial_MSE)
plt.title('MSE a tráves de las iteraciones')
plt.xlabel('Iteración')
plt.ylabel('MSE')
plt.show()
```
