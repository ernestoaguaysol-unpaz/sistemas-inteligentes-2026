```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Carga del dataset y creación de X e y
dataset = pd.read_csv('dataset_vinos.csv')
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
# Separación en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Separación en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
# Podemos ver qué forma (dimensiones) tienen los conjuntos -> cantidad de ejemplos de c/u.
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Ahora para fit usamos X_train, y_train
modelo = DecisionTreeClassifier(max_depth=3)
modelo.fit(X_train, y_train)
# Ver la estructura del árbol aprendido
plot_tree(modelo)
```

```python
# Midiendo la exactitud con el conjunto de prueba
accuracy_score(y_test, modelo.predict(X_test))
```
