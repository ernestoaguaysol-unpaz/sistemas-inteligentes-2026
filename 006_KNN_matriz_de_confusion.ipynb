```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Carga del dataset y creación de X e y
dataset = pd.read_csv('dataset_vinos.csv')
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]

# Separación en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
# Podemos ver qué forma (dimensiones) tienen los conjuntos -> cantidad de ejemplos de c/u.
X_train.shape, X_test.shape, y_train.shape, y_test.shape

# Ahora para fit usamos X_train, y_train
modelo = KNeighborsClassifier(n_neighbors=5)
modelo.fit(X_train, y_train)

# Midiendo la exactitud con el conjunto de prueba
accuracy_score(y_test, modelo.predict(X_test))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Matriz de confusión y exactitud para datos de entrenamiento
ConfusionMatrixDisplay.from_estimator(modelo,
   X_train,
   y_train)
plt.show()
print('Exactitud con datos de entrenamiento:’,
      accuracy_score(y_train, modelo.predict(X_train)))

```

```python
# Matriz de confusión y exactitud para datos de prueba
ConfusionMatrixDisplay.from_estimator(modelo,
   X_test,
   y_test)
plt.show()
print('Exactitud con datos de entrenamiento:’,
      accuracy_score(y_test, modelo.predict(X_test)))
```
