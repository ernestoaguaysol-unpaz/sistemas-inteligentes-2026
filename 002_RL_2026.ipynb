```python
# Librerías necesarias
import pandas as pd
from sklearn.linear_model import LinearRegression
# Matplotlib: visualizaciones de datos
# https://matplotlib.org/
import matplotlib.pyplot as plt
```

```python
# Carga del dataset y creación de X e y
dataset = pd.read_csv('dataset_hs_jugadas_puntaje.csv')
X = dataset.iloc[:, :-1]# Todas las filas, todas las columnas menos la última
y = dataset.iloc[:, -1] # Todas las filas, última columna
```

```python
# Vamos a inspeccionar los primeros datos usando el método head(filas=5)
dataset.head()
```

```python
dataset.plot.scatter(x='HsJuego', y='Puntaje')
#  plt.show()
```

```python
# Crear H - objeto de la clase base: Regresión lineal. Modelo sin ajustar / sin entrenar.
modelo = LinearRegression()
```

```python
# Ajustar H → crear h → ajustar/entrenar el modelo
modelo.fit(X, y)
```

```python
# Usarlo para hacer inferencia -realizar predicciones- con nuevos datos.
modelo.predict(
   pd.DataFrame([
       [6],
   ], columns=['HsJuego'])
)
```
