```python
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid")
from sklearn.neighbors import KNeighborsClassifier
```

```python
dataset = pd.read_csv('dataset_vinos.csv')
dataset.head()
```

```python
sns.scatterplot(data=dataset, x='alcohol', y='intensidad_color', hue='clase')
```

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
# Crear H - objeto de la clase base: KNN con K=5. Modelo sin ajustar / sin entrenar.
modelo = KNeighborsClassifier(n_neighbors=5)
```

```python
# Ajustar H → crear h → ajustar/entrenar el modelo
modelo.fit(X, y)
```

```python
# Usarlo para hacer inferencia -realizar predicciones- con nuevos datos.
modelo.predict(	# predict -> pred de la clase | predict_proba -> pred de las probs.
   pd.DataFrame([
       [12, 6],
   ], columns=['alcohol','intensidad_color'])
)
```

```python
modelo.predict_proba(	# predict -> pred de la clase | predict_proba -> pred de las probs.
   pd.DataFrame([
       [12, 6],
   ], columns=['alcohol','intensidad_color'])
)
```
